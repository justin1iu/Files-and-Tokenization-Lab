        When I tokenizd, I removed all punctuation, which meant that some words would be strangely chopped in halfâ€”for example, o'er becomes oer, which is not a real word. Through experimentation, I realized that keeping capital letters would make the list way too complicated, and to refine that, I converted all tokens to full lowercase via the .lower() method. 



How many tokens does your tokenization method obtain? How many types?
        Using the len() function at the end of my code, I got 1812 unique tokens (types) in my list and 6452 total tokens. I went from non-unique to unique by using the Counter method to sum up the same tokens. 

```python
from pathlib import Path
from collections import Counter
import re

file = Path("book_9.txt")

with file.open("r") as file:
    text = file.read().lower()

clean_text = re.sub(r'[^a-z\s]', '', text)
tokens = clean_text.split()

token_counts = Counter(tokens)

for token, count in token_counts.items():
    print(f"{token}: {count}")


print(f"There are {len(token_counts)} unique tokens in the tokens list, and {len(tokens)} non-unique tokens.")
```
